# Deal with missing data
How to deal with missing data in this case?
drop data
a. drop the whole row of price as it is the target column.
replace data
a. replace it by mean
1/ "normalized-losses": 41 missing data, replace them with mean
2/ "stroke": 4 missing data, replace them with mean
3/ "bore": 4 missing data, replace them with mean
4/ "horsepower": 2 missing data, replace them with mean
5/ "peak-rpm": 2 missing data, replace them with mean

b. replace it by frequency
-- "num-of-doors": 2 missing data, replace them with "four".
Reason: 84% sedans is four doors. Since four doors is most frequent, it is most likely to occur

```python
avg_norm_loss = df['normalized-losses'].astype(float).mean()
print("Average of normalized-losses:", avg_norm_loss)
```

```python
# Raplace null by average value:
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
avg_bore=df['bore'].astype('float').mean()
print("Average of bore:", avg_bore)
```

```python
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
avg_stroke = df['stroke'].astype('float').mean()
print("Average stroke:", avg_stroke)
```

```python
df["stroke"].replace(np.nan, avg_bore, inplace=True)
```

```python
avg_horsepower = df['horsepower'].astype(float).mean()
print("Average horsepower:", avg_horsepower)
```

```python
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
avg_peakrpm=df['peak-rpm'].astype('float').mean()
print("Average peak rpm:", avg_peakrpm)
```

```python
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
# Four is a more common value as most cars also have four doors.
# Hence, we replace missing values by 'four'
```

```python
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
# Now, drop all the row with missing data in 'price' column:
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

# Correct data format:

```python
df.dtypes
```

```python
# 'bore' and 'stroke' variables are numerical so their types should be integer or float
# 'price', 'peak-rpm', 'normalized-lossed' should be 'float' type too.
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

# Data standardization

```python
df['city-L/100km'] = 235/df["city-mpg"]
```

```python
df.rename(columns={'city-L/100km':'highway-L/100'},inplace=True)
df.head()
```

# Data Normalization

```python
# replace (original value) by (original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
#Here we can see, we've normalized "length", "width" in the range of [0,1]. 
```

```python
#Standardize with z-score
df['height'] = (df['height']-df['height'].mean())/df['height'].std()
df['height']
```

```python
df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
import seaborn as sns
sns.set_style(style='darkgrid')
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# Group horsepower into three groups:
group_names = ['Low', 'Medium', 'High']
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head()
```

```python
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```
